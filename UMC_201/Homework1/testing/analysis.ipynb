```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("results_with_index.csv")
```

```python
data.head()
```

```python
data.plot(x='Size', y='Py_Time', kind='line')
plt.xlabel('Size (n)')
plt.ylabel('Py_Time (s)')
plt.title('Size vs Py_Time')
plt.show()
```

```python
data.plot(x='Size', y='Cpp_Time', kind='line')
plt.xlabel('Size (n)')
plt.ylabel('Cpp_Time (s)')
plt.title('Size vs Cpp_Time')
plt.show()
```

```python
data.plot(x='Size', y=['Py_Time', 'Cpp_Time'], kind='line')
plt.xlabel('Size')
plt.ylabel('Time (s)')
plt.title('Size vs Time')
plt.legend(['Py_Time', 'Cpp_Time'])
plt.show()
```
